```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import cumsum
data1=[1007,1194,1198,1309,1354,1412,1679,2198,2727,4500]
data2=[1003,1167,1161,1250,1295,1342,1603,1803,2001,2500]
def calcdf(data):
    data_size=len(data)
        # Set bins edges
    data_set=sorted(set(data))
    bins=np.append(data_set, data_set[-1]+1)
        # Use the histogram function to bin the data
    counts, bin_edges = np.histogram(data, bins=bins, density=False)
    counts=counts.astype(float)/data_size
            # Find the cdf
    cdf = np.cumsum(counts)

    plt.ylim((0,1))
    plt.ylabel("CDF")
   # plt.plot(bin_edges[0:-1], cdf,'r-')
    plt.xlabel("Shuffle Finish Time (seconds)")
    plt.grid(True)
    return bin_edges[0:-1],cdf
    # Plot the cdf

x, cdf=calcdf(data1)
plt.plot(x, cdf,'r-')
y, cdf=calcdf(data2)
plt.plot(y, cdf,'g-')
plt.show()
x
```
